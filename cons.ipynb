```python
import requests
import os
from datetime import datetime

class ChatExporter:
    def _init_(self, base_url="http://localhost:5000"):
        self.base_url = base_url.rstrip('/')
        self.download_folder = "chat_exports"
        
        # Crear carpeta de descargas si no existe
        if not os.path.exists(self.download_folder):
            os.makedirs(self.download_folder)

    def download_csv(self, endpoint, filename_prefix):
        """
        Descarga un archivo CSV desde el endpoint especificado
        """
        try:
            # Realizar la petición al endpoint
            response = requests.get(f"{self.base_url}/export/{endpoint}")
            
            # Verificar si la petición fue exitosa
            if response.status_code == 200:
                # Crear nombre de archivo con timestamp
                timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
                filename = f"{filename_prefix}_{timestamp}.csv"
                filepath = os.path.join(self.download_folder, filename)
                
                # Guardar el archivo
                with open(filepath, 'wb') as f:
                    f.write(response.content)
                    
                print(f"Archivo descargado exitosamente: {filepath}")
                return filepath
            else:
                print(f"Error al descargar {endpoint}: {response.status_code}")
                print(f"Respuesta: {response.text}")
                return None
                
        except Exception as e:
            print(f"Error durante la descarga de {endpoint}: {str(e)}")
            return None

    def export_all(self):
        """
        Exporta tanto los chats como los detalles de chat
        """
        print("Iniciando exportación de datos...")
        
        # Exportar chats
        chats_file = self.download_csv('chats', 'chats_export')
        if chats_file:
            print(f"Chats exportados correctamente")
        
        # Exportar detalles de chat
        details_file = self.download_csv('chat_details', 'chat_details_export')
        if details_file:
            print(f"Detalles de chat exportados correctamente")
            
        return chats_file, details_file

def main():
    # URL de tu API
    api_url = "http://localhost:5000"  # Cambia esto según tu configuración
    
    # Crear instancia del exportador
    exporter = ChatExporter(api_url)
    
    # Realizar la exportación
    print(f"Conectando a {api_url}...")
    chats_file, details_file = exporter.export_all()
    
    if chats_file and details_file:
        print("\nExportación completada exitosamente!")
        print(f"Archivos generados:")
        print(f"- Chats: {chats_file}")
        print(f"- Detalles: {details_file}")
    else:
        print("\nLa exportación no se completó correctamente")

if _name_ == "_main_":
    main()
```
